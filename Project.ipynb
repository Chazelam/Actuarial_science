```python
import pandas as pd
pd.options.mode.chained_assignment = None
```

```python
# Постоянные:
quantile = 1.645
unit = 1000

df = pd.ExcelFile("data.xlsx")
df_1 = df.parse('Слесарь 4 разряд', index_col=0).drop(['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 6', 'Ri*'], axis=1).dropna(subset='Pi*')
df_1['Pi*'] = df_1['Pi*'].astype(float)
N1 = 50

df_2 = df.parse('Электрогазосварщик 5 разряда', index_col=0).drop(['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 6', 'Ri*'], axis=1).dropna(subset='Pi*')
df_2['Pi*'] = df_2['Pi*'].astype(float)
N2 = 100

df_3 = df.parse('Электрогазосварщик 4 разряда', index_col=0).drop(['Unnamed: 1', 'Unnamed: 3', 'Unnamed: 5', 'Unnamed: 6', 'Ri*'], axis=1).dropna(subset='Pi*')
df_3['Pi*'] = df_3['Pi*'].astype(float)
N3 = 300

df_1[:3]
```

```python
df_1.loc[df_1['Pi*'] == 1, "Pi*"] = round(1/365, 4)
df_1.loc[df_1['Pi*'] == 3, "Pi*"] = round(4/365, 4)
df_1.loc[df_1['Pi*'] == 7, "Pi*"] = round(12/365, 4)

df_2.loc[df_2['Pi*'] == 1, "Pi*"] = round(1/365, 4)
df_2.loc[df_2['Pi*'] == 3, "Pi*"] = round(4/365, 4)
df_2.loc[df_2['Pi*'] == 7, "Pi*"] = round(12/365, 4)

df_3.loc[df_3['Pi*'] == 1, "Pi*"] = round(1/365, 4)
df_3.loc[df_3['Pi*'] == 3, "Pi*"] = round(4/365, 4)
df_3.loc[df_3['Pi*'] == 7, "Pi*"] = round(12/365, 4)

df_1[:3]
```

```python
value = df_1['Ui*']
prob = df_1["Pi*"]
Ex1 = sum([value[i]*prob[i] for i in range(len(df_1))])
E2x1 = sum([(value[i]**2)*prob[i] for i in range(len(df_1))])
VarX1 = E2x1 - Ex1**2

value = df_2['Ui*']
prob = df_2["Pi*"]
Ex2 = sum([value[i]*prob[i] for i in range(len(df_2))])
E2x2 = sum([(value[i]**2)*prob[i] for i in range(len(df_2))])
VarX2 = E2x2 - Ex2**2

value = df_3['Ui*']
prob = df_3["Pi*"]
Ex3 = sum([value[i]*prob[i] for i in range(len(df_3))])
E2x3 = sum([(value[i]**2)*prob[i] for i in range(len(df_3))])
VarX3 = E2x3 - Ex3**2

print(Ex1, VarX1)
VarS = VarX1*N1 + VarX2*N2 + VarX3*N3
Es = N1*Ex1 + N2*Ex2 + N3*Ex3
```

```python
U = Es + quantile*VarS**(1/2)
O = (quantile*VarS**(1/2))/Es
Net1 = Ex1*(1 + O)
Net2 = Ex2*(1 + O)
Net3 = Ex3*(1 + O)
```

```python
print(f'Стартовый капитал: {round(U*unit, 2)} р.')
print('Нетто-премия:')
print(f'   {df.sheet_names[0]}: {round(Net1*unit, 2)} р.')
print(f'   {df.sheet_names[1]}: {round(Net2*unit, 2)} р.')
print(f'   {df.sheet_names[2]}: {round(Net3*unit, 2)} р.')
```
